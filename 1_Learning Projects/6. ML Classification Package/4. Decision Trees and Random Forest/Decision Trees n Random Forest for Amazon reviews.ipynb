DATA CLEANINIG AND FEATURE ENGINEERING

```python
data_Amazon.drop(["date","rating"] , axis = 1, inplace = True)
```

```python
data_Amazon
```

```python
variation_dummies = pd.get_dummies(data_Amazon.variation, drop_first=True)
variation_dummies
```

```python
variation_dummies.columns
```

```python
data_Amazon.drop(["variation"], axis = 1, inplace = True)
```

```python
data_Amazon = pd.concat([data_Amazon, variation_dummies], axis = 1)
```

```python
data_Amazon 
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
data_countVector = vectorizer.fit_transform(data_Amazon.verified_reviews) # Used to transform the bunch of strings into numerals by simply creating columns for each String value
```

```python
data_countVector.shape
```

```python
print(vectorizer.get_feature_names())
```

```python
print(data_countVector.toarray())
```

```python
data_Amazon.drop(["verified_reviews"], axis = 1, inplace = True)
```

```python
encoded_Reviews = pd.DataFrame(data_countVector.toarray())
```

```python
X = pd.concat([encoded_Reviews,data_Amazon], axis = 1).copy()
```

```python
X 
```

```python
Y = X["feedback"].copy()
```

```python
X.drop(["feedback"], axis =1, inplace = True)
```

```python
Y
```

```python
X
```

TRAINING THE MODEL

```python
from sklearn.model_selection import train_test_split
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size = 0.2, random_state = 10)
```

```python
X_Train.shape
```

```python
X_Test.shape
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
randomForestClassifier = RandomForestClassifier(n_estimators=200, criterion = "entropy")
randomForestClassifier.fit(X_Train, Y_Train)
```

Model Evaluation 


```python
Y_predict_Train = randomForestClassifier.predict(X_Train)
Y_predict_Train
```

```python
cm = confusion_matrix(Y_Train, Y_predict_Train)
akws = {"ha": 'center',"va": 'center'}
sns.heatmap(cm, annot = True, annot_kws=akws)
```

```python
print(classification_report(Y_Train, Y_predict_Train))
```

```python
Y_predict_Test = randomForestClassifier.predict(X_Test)
Y_predict_Test
```

```python
cm = confusion_matrix(Y_Test, Y_predict_Test)
akws = {"ha": 'center',"va": 'center'}
sns.heatmap(cm, annot = True, annot_kws=akws)
```

```python
print(classification_report(Y_Test, Y_predict_Test))
```

IMPROVE THE MODEL 

```python
data_Amazon = pd.read_csv("amazon_alexa.tsv", sep="\t")

data_countVector = vectorizer.fit_transform(data_Amazon.verified_reviews)
data_Amazon.columns
```

```python
data_Amazon = pd.concat([data_Amazon, pd.DataFrame(data_countVector.toarray())], axis = 1)
```

```python
data_Amazon
```

```python
data_Amazon["length_reviews"] = data_Amazon.verified_reviews.apply(len)
```

```python
data_Amazon.length_reviews
```

```python
X = data_Amazon.drop(['rating', 'date', 'variation', 'verified_reviews', 'feedback'], axis =1)
```

```python
Y = data_Amazon.feedback.copy()
```

```python
from sklearn.model_selection import train_test_split
X_Train, X_Test, Y_Train, Y_Test = train_test_split(X, Y, test_size = 0.2, random_state = 10)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.ensemble import RandomForestClassifier
randomForestClassifier = RandomForestClassifier(n_estimators=200, criterion = "entropy")
randomForestClassifier.fit(X_Train, Y_Train)
```

```python
Y_predict_Test = randomForestClassifier.predict(X_Test)
Y_predict_Test
```

```python
cm = confusion_matrix(Y_Test, Y_predict_Test)
sns.heatmap(cm, annot = True)
```

```python
print(classification_report(Y_Test, Y_predict_Test))
```
